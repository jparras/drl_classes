We assume that we know:
* The transition probabilities: $\mathcal{P}_{xx}^{u}=0.8$, $\mathcal{P}_{xy}^{u}=0.2$, $\mathcal{P}_{xx}^{m}=0.2$, $\mathcal{P}_{xy}^{m}=0.8$, $\mathcal{P}_{yx}^{u}=0.3$, $\mathcal{P}_{yy}^{u}=0.7$, $\mathcal{P}_{yx}^{m}=0.9$, $\quad\mathcal{P}_{yy}^{m}=0.1$.
* The discount factor is $\gamma=0.9$.
* The rewards are: $\mathcal{R}_{x}^{u}=-1$, $\mathcal{R}_{x}^{m}=0.6$,
$\mathcal{R}_{y}^{u}=0.5$, $\mathcal{R}_{y}^{m}=-0.9$.

**You have to solve the following four points**:
* Write the value functions in vector form: $v^{\pi}\triangleq\left(v^{\pi}(s)\right)_{s\in\mathcal{S}}$ and $q^{\pi}\triangleq\left(q^{\pi}(s,a)\right)_{s\in\mathcal{S},a\in\mathcal{A}}$.

* Write the expected reward vector, the transition probability matrix and the policy in matrix form.

* Write the transition probability matrix and the expected reward vector for all possible deterministic policies.

* Obtain the state and state-action value functions using the matrix form for each of the deterministic policies.

The next cells contain the solutions to these exercises, so try to solve them before advancing!

Regarding the **first point**: we need to write the value functions, which means that we only need to pay attention to the number of states and actions that we have. Either from the figure or the probability and reward information, we know that $\mathcal{S}=\{x, y\}$, so $|\mathcal{S}|=2$, and $\mathcal{A}=\{u, m\}$, so $|\mathcal{A}|=2$. Hence, the solution is:

$v^{\pi}=\left[\begin{array}{c}
v^{\pi}\left(x\right)\\
v^{\pi}\left(y\right)
\end{array}\right]\in\mathbb{R}^{|\mathcal{S}|}=\mathbb{R}^{2},\quad q^{\pi}=\left[\begin{array}{c}
q^{\pi}\left(x,u\right)\\
q^{\pi}\left(x,m\right)\\
q^{\pi}\left(y,u\right)\\
q^{\pi}\left(y,m\right)
\end{array}\right]\in\mathbb{R}^{|\mathcal{S}||\mathcal{A}|}=\mathbb{R}^4$

Regarding the **second point**, the solution relies on the data given and is:
* The reward vector is: $\mathcal{R}=\left[\begin{array}{c}
\mathcal{R}_{x}^{u}\\
\mathcal{R}_{x}^{m}\\
\mathcal{R}_{y}^{u}\\
\mathcal{R}_{y}^{m}
\end{array}\right]\in\mathbb{R}^4$.

* The transition probability matrix is: $\mathcal{P}=\left[\begin{array}{cc}
\mathcal{P}_{xx}^{u} & \mathcal{P}_{xy}^{u}\\
\mathcal{P}_{xx}^{m} & \mathcal{P}_{xy}^{m}\\
\mathcal{P}_{yx}^{u} & \mathcal{P}_{yy}^{u}\\
\mathcal{P}_{yx}^{m} & \mathcal{P}_{yx}^{m}
\end{array}\right]\in\mathbb{R}^{4 \times 2}$.

* The policy in matrix form is: $\Pi=\left[\begin{array}{cccc}
\pi\left(u\mid x\right) & \pi\left(m\mid x\right) & 0 & 0\\
0 & 0 & \mathbb{\pi}\left(u\mid y\right) & \pi\left(m\mid y\right)
\end{array}\right]\in\mathbb{R}^{2 \times 4}$.

Regarding the **third point**, we recall the following formulations from the theory:

* $\mathcal{P}_{ss\text{'}}^{\pi}=	\sum_{a\in\mathcal{A}}\pi\left(a|s\right)\mathcal{P}_{ss'}^{a}\quad\rightarrow\quad\mathcal{P}^{\pi}=\mathit{\varPi}\mathcal{P}$

* $\mathcal{R}_{s}^{\pi}=	\sum_{a\in\mathcal{A}}\pi\left(a|s\right)\mathcal{R}_{s}^{a}\quad\rightarrow\quad\mathcal{R}^{\pi}=\varPi\mathcal{R}$

We have four possible deterministic policies, which yield the following transition matrices and reward vectors:
* $\varPi_{1}=\left[\begin{array}{cccc}
1 & 0 & 0 & 0\\
0 & 0 & 1 & 0
\end{array}\right]$,
$\mathcal{P}^{\pi_{1}}=\left[\begin{array}{cc}

\end{array}\right]$,
$\mathcal{R}^{\pi_{1}}=\left[\begin{array}{c}

* $\varPi_{2}=\left[\begin{array}{cccc}

0 & 0 & 1 & 0
\end{array}\right]$,
$\mathcal{P}^{\pi_{2}}=\left[\begin{array}{cc}
0.2 & 0.8\\
0.3 & 0.7
\end{array}\right]$,
$\mathcal{R}^{\pi_{2}}=\left[\begin{array}{c}
0.6\\
0.5
\end{array}\right]$.

* $\varPi_{3}=\left[\begin{array}{cccc}
1 & 0 & 0 & 0\\
0 & 0 & 0 & 1
\end{array}\right]$,
$\mathcal{P}^{\pi_{3}}=\left[\begin{array}{cc}
0.8 & 0.2\\
0.9 & 0.1
\end{array}\right]$,
$\mathcal{R}^{\pi_{3}}=\left[\begin{array}{c}

* $\varPi_{4}=\left[\begin{array}{cccc}
0 & 1 & 0 & 0\\

$\mathcal{P}^{\pi_{4}}=\left[\begin{array}{cc}

Let us now code these results, in order to check them. First, we import the usual libraries:

```python
import numpy as np
from tabulate import tabulate
```

And then, we proceed to encode the information given on the problem, as well as the four policy matrices that we have just defined. We also output the results we have just presented, in order to cross-check their validity.

```python
# Define here the matrices given by the problem
P = np.array([[0.8, 0.2], [0.2, 0.8], [0.3, 0.7], [0.9, 0.1]])
R = np.array([[-1], [0.6], [0.5], [-0.9]])
gamma = 0.9
# Now, define all possible policy matrices
Pi = [np.zeros((2, 4)) for _ in range(4)]
Pi[0][0, 0] = 1
Pi[0][1, 2] = 1
Pi[1][0, 1] = 1
Pi[1][1, 2] = 1
Pi[2][0, 0] = 1
Pi[2][1, 3] = 1
Pi[3][0, 1] = 1
Pi[3][1, 3] = 1

for i, p in enumerate(Pi):
  print('Results for policy ' + str(i + 1) + ':')
  print('Policy')
  print(p)
  print('Probability transition matrix')
  print(p @ P)
  print('Reward vector')
  print(p @ R)
```

Regarding the **fourth point**, we first start with the state-value function, whose expression, in matrix form, is:

$v^{\pi}=\left(I-\gamma\mathcal{P^{\pi}}\right)^{-1}\mathcal{R}^{\pi}$

We implement this expression in the code in order to obtain the following results:

$v^{\pi_{1}}=\left[\begin{array}{c}
-5.09\\
-2.36
\end{array}\right],v^{\pi_{2}}=\left[\begin{array}{c}
5.34\\
5.25
\end{array}\right],v^{\pi_{3}}=\left[\begin{array}{c}
-9.84\\
-9.74
\end{array}\right],v^{\pi_{4}}=\left[\begin{array}{c}
-0.63\\
-1.54
\end{array}\right]$

```python
# Compute the state value function for each policy
v_pi = []
for i, pi in enumerate(Pi):
  P_pi = pi @ P
  R_pi = pi @ R
  v_pi.append([i + 1])
  v_pi[i].extend(list(np.linalg.inv(np.eye(P_pi.shape[0]) - gamma * P_pi) @ R_pi))

print('State value functions obtained for each policy (Policy Evaluation)')
print(tabulate(v_pi, tablefmt="fancy_grid", headers=['Policy', 'V(x)', 'V(y)']))
```

Regarding the state-action value function, its expression in matrix form is:

$q^{\pi}=\left(I-\gamma\mathcal{P}\varPi\right)^{-1}\mathcal{R}$

```python
# Compute the state-action value function for each policy
q_pi = []
for i, pi in enumerate(Pi):
  q_pi.append([i + 1])
  q_pi[i].extend(list(np.linalg.inv(np.eye(P.shape[0]) - gamma * P @ pi) @ R))

print('State-action value functions obtained for each policy (Policy Evaluation)')
print(tabulate(q_pi, tablefmt="fancy_grid", headers=['Policy', 'q(x,u)', 'q(x,m)', 'q(y,u)', 'q(y,m)']))
```
